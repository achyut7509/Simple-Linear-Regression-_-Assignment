# Dataset results using model transformation :

## 1. Importing Libraries :

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

## 2. Importing dataset :

```python
salary_original_dataset = pd.read_csv('Salary_Data.csv',sep=',')
salary_original_dataset
```

### 3. Data Preparation :

```python
# Renaming columns to build our model
salary_original_dataset.columns = ['years_experience','salary']
salary_original_dataset.head()
```

### 4. Analysis using data transforation :

### Model -1 :

```python
import statsmodels.formula.api as smf
salary_hike_linear_model_1 = smf.ols(formula = 'salary ~ years_experience',data = salary_original_dataset).fit()
salary_hike_linear_model_1
```

```python
print('R2 Score             :',salary_hike_linear_model_1.rsquared)
print('R2 adjusted Score    :',salary_hike_linear_model_1.rsquared_adj)
print('AIC                  :',salary_hike_linear_model_1.aic)
print('BIC                  :',salary_hike_linear_model_1.bic)
```

```python
salary_hike_linear_model_1.summary()
```

### Model-2  :

```python
salary_hike_dataset_1 = salary_original_dataset.copy()
salary_hike_dataset_1.head()
```

```python
salary_hike_dataset_1['log_salary'] = np.log(salary_hike_dataset_1['salary'])
salary_hike_dataset_1.head()
```

```python
salary_hike_dataset_1.drop(labels=['salary'],axis=1,inplace=True)
salary_hike_dataset_1.head()
```

```python
import statsmodels.formula.api as smf
salary_hike_linear_model_2 = smf.ols(formula = 'log_salary ~ years_experience',data = salary_hike_dataset_1).fit()
salary_hike_linear_model_2
```

```python
print('R2 Score             :',salary_hike_linear_model_2.rsquared)
print('R2 adjusted Score    :',salary_hike_linear_model_2.rsquared_adj)
print('AIC                  :',salary_hike_linear_model_2.aic)
print('BIC                  :',salary_hike_linear_model_2.bic)
```

```python
salary_hike_linear_model_2.summary()
```

### Model-4 :

```python
salary_hike_dataset_3 = salary_original_dataset.copy()
salary_hike_dataset_3.head()
```
